```python
import sqlite3
import datetime
from lib import get_movie_infos
```

```python
def add_movie(res, seen=0):
    conn = sqlite3.connect('test.db')
    conn.execute('''INSERT INTO MOVIES (actors, date, directors, duration, genres, poster_link,
                                    press_rating, seen, spec_rating, synopsis, title)
                VALUES (?,?,?,?,?,?,?,?,?,?,?)''',
            (';'.join(res['actors']), 
             res['date'], 
             ';'.join(res['directors']),
             int(res['duration'].total_seconds()),
             ';'.join(res['genres']),
             res['poster_link'],
             res['ratings']['press'],
             seen,
             res['ratings']['spectators'],
             res['synopsis'],
             res['title']
            ))
```

# Create Database

```python
conn = sqlite3.connect('test.db')
```

```python
conn.execute('DROP TABLE IF EXISTS MOVIES;')
conn.execute('''CREATE TABLE MOVIES
        (id INTEGER PRIMARY KEY AUTOINCREMENT,
         actors TEXT,
         date DATE NOT NULL,
         directors TEXT,
         duration INT,
         genres TEXT,
         poster_link TEXT,
         press_rating NUMERIC,
         seen BOOLEAN DEFAULT 0,
         spec_rating NUMERIC,
         synopsis TEXT,
         title TEXT NOT NULL,
         video_link TEXT
         )''')
```
